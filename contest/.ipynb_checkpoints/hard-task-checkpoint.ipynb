```python
import pandas as pd
import numpy as np
```

```python
events_data = pd.read_csv('data/event_data_train.zip')
submission_data = pd.read_csv('data/submissions_data_train.zip')

submission_data.head()
```

```python
# задача - найти степ-гроб. Степ, после которого наибольшая часть 
# пользователей больше ничего не сабмитила и не пыталась

# 1. выбираем все степы у которых сабмит wrong и берем из них максимальные таймстемпы
df = submission_data[submission_data['submission_status']=='wrong' ]

df = df.groupby('user_id', as_index=False)\
  .agg({'timestamp':'max'})\
  .rename(columns={'timestamp':'last_wrong_timestamp'})
df.head()
```

```python
# inner - потому что нам нужны только те, у которых есть хоть один степ
df_mix = pd.merge(submission_data, df, on='user_id', how='inner',
                  left_index=False, right_index=False, sort=False,
                  suffixes=('_x', '_y'), copy=True, indicator=False,
                  validate=None)
df_mix_last_wrong = df_mix[ (df_mix['last_wrong_timestamp'] == df_mix['timestamp']) 
                          & (df_mix['submission_status'] == 'wrong')]
df_mix_last_wrong
```
