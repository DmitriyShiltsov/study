```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'a':[1,2,1,2,3,3,4,3,2,1,3,4, 6], 'b' : [10,20,10,20,30,30,40,30,20,10,30,40, 60]})
df2 = pd.DataFrame({'a':[1,2,3,4], 'val':['a1','a2','a3','a4']})
```

```python
df1
```

```python
pd.merge(df1, df2, how='left', left_on='a', right_on='a',
         left_index=False, right_index=False, sort=False,
         suffixes=('_x', '_y'), copy=True, indicator=False,
         validate=None)
```
